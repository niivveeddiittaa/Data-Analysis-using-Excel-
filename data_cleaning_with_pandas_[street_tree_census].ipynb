# **Lets clean our dataset**

```python
mask = ((df_subset['status']=='Stump') | (df_subset['status']=='Dead'))
```

```python
df_subset.loc[mask] = df_subset.loc[mask].fillna('not applicable')
```

```python
df_subset[df_subset['status']=='Stump']
```

```python
df_subset.isna().sum()
```

# **remove ALL THE NULL VALS**

```python
df_subset.fillna({'health': 'Good'}, inplace=True)
df_subset.fillna({'steward': 'No steward'}, inplace=True)
df_subset.fillna({'problems': 'No problems'}, inplace=True)
df_subset
```

```python
df_subset.isna().sum()
```

# **dealing with OUTLIERS**

```python
bigtree = df_subset[(df_subset['tree_dbh'] <= 60) & (df_subset['stump_diam'] <= 60)]
bigtree

```

```python
df_alive_trees = df_subset[df_subset['status']=='Alive']
df_dead_or_stump_trees = df_subset[(df_subset['status']=='Dead') | (df_subset['status']=='Stump')]
```

```python
df_alive_trees.groupby('spc_latin')['tree_dbh'].describe()
```

# **dealing with unexpected data issues**

```python
# table created

stats_alive = df_alive_trees.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin','25%', '75%']]
stats_alive
```

```python
df_alive_trees.merge(stats_alive, on='spc_latin', how='left')
df_alive_trees
```
