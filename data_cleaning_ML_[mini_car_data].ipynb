# **extracting company name**

# **if price column is in "*28 Lakh*" or "*13 crore*" type**

```python
def rupee_change(x):
    p = x.split(" ")
    try :
        if p[1] == "Lakh":
            return str(round(float(p[0])*100000,1))
        elif p[1] == "crore":
            return str(round(float(p[0])*10000000,1))
    except:
        return x
```

```python
df['car_price_rs'] = df['car_price'].apply(rupee_change)
df['car_price_rs'] = df['car_price_rs'].astype(float)
```

```python
df.info()
```

# **finding out "*inter quartile range*" to removing outliersss**

```python
q1 = np.quantile(df["engine"],0.25)
q3 = np.quantile(df["engine"],0.75)
```

```python
iqr = q3 - q1
iqr
```

```python
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
```

```python
min_r,max_r
```

```python
df = df[df["engine"]<=max_r]
```
