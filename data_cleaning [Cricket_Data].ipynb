# **rename column_names**

```python
df = df.rename(columns={'NO':"Not_Outs" , 'HS':'Highest_Inns_Score' , 'BF':'Balls_Faced' , 'SR':'Strike_Rate' , 'Mat':'Matches' , 'Ave':'Average'})
df
```

# **which row has null values**

```python
df[df['Balls_Faced'].isna()==1]
```

# **removing null values**

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df['Strike_Rate'] = df['Strike_Rate'].fillna(0)
```

# **if we want to see a particular row**

```python
df[df['Player']=='ED Weekes (WI)']
```

# **dropping duplicates**

```python
df[df['Player'].duplicated()]
```

```python
# see exactly how many duplicate rows are present
df[df['Player'].isin(['GA Headley (WI)','GS Sobers (WI)','JH Kallis (ICC/SA)','V Sehwag (ICC/IND)'])]
```

```python
df = df.drop_duplicates()
```

```python
df[df['Player'].isin(['GA Headley (WI)','GS Sobers (WI)','JH Kallis (ICC/SA)','V Sehwag (ICC/IND)'])]
```

# **splitting span into start date and end date**

```python
df['Span'].str.split(pat ='-')
```

```python
df['Start_Date'] = df['Span'].str.split(pat ='-').str[0]
```

```python
df['End_Date'] = df['Span'].str.split(pat ='-').str[1]
```

```python
df
```

```python
df = df.drop(columns=['Span'],axis=1)
```

```python
df
```

# **splitting player into Player and country**

```python
df['Player'].str.split(pat ='(')
```

# **removing () from country**

```python
df['Country'] = df['Player'].str.split(pat ='(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat =')').str[0]
```

```python
df['Country']
```

# **keeping only player names**

```python
df['Player'] = df['Player'].str.split(pat ='(').str[0]
```

```python
df['Player']
```

```python
df
```

# **changing data types**

```python
df.dtypes
```

# **"Highest_Inns_Score" column**

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat ='*').str[0]
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
```

```python
df.dtypes
```

# **"Start_Date" and "End_Date" columns**

```python
df = df.astype({'Start_Date':'int' , 'End_Date':'int' })
```

```python
df.dtypes
```

# **"0" column**

# **build "career duration" column**

```python
df['Career_Duration'] = df['End_Date'] - df['Start_Date']
```

```python
df
```

# **ANALYSIS**
# **Average careers wrt players**

```python
df['Career_Duration'].mean()
```

# **average strike rate for cricketers who played more than 10 years**

```python
df[df['Career_Duration'] > 10]['Strike_Rate'].mean()
```

# **number of cricketers who played before 1960**

```python
df[df['Start_Date'] < 1960]['Player'].count()
```

# **max highest inns scored by country**

```python
df.groupby('Country')['Highest_Inns_Score'].max().to_frame('Highest_Inns_Score').reset_index().sort_values('Highest_Inns_Score',ascending=True)
```

# **hundreda, 50s, 100s, 0s avg by country**

```python
df.groupby('Country')[['100','50','0']].mean()
```
